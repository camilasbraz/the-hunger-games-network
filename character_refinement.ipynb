```python
import pandas as pd
import re
```

```python
character_df = pd.read_csv("characters.csv")
character_df 
```

```python
# Remove brackets and text within brackets
character_df['character'] = character_df['character'].apply(lambda x: re.sub("[\(].*?[\)]", "", x)) 
character_df['character_firstname'] = character_df['character'].apply(lambda x: x.split(' ', 1)[0])
character_df
```

```python
nickname_mapping = {
    'Primrose': 'Prim',
}

character_df['character_nickname'] = character_df['character_firstname'].map(nickname_mapping)
```

```python
character_df.to_csv('characters_ref.csv', index=False)
```
